```python
import torchaudio
from speechbrain.pretrained import Tacotron2, HIFIGAN
from IPython.display import Audio  # For playing audio in runtime

# Initialize TTS (Tacotron2) and Vocoder (HiFi-GAN)
tacotron2 = Tacotron2.from_hparams(source="speechbrain/tts-tacotron2-ljspeech", savedir="tmpdir_tts")
hifi_gan = HIFIGAN.from_hparams(source="speechbrain/tts-hifigan-ljspeech", savedir="tmpdir_vocoder")

# Running the TTS
text = "the weather is bad"
mel_output, mel_length, alignment = tacotron2.encode_text(text)

# Running Vocoder (spectrogram-to-waveform)
waveforms = hifi_gan.decode_batch(mel_output)

# Save the waveform to a file (optional)
output_path = 'example_TTS.wav'
torchaudio.save(output_path, waveforms.squeeze(1), 22050)
print(f"Audio saved to {output_path}")

# Display and play the audio in runtime
print("Playing the synthesized audio:")
Audio(data=waveforms.squeeze(1).numpy(), rate=22050)
```
